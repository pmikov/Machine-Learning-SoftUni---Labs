```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris, load_boston

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, AdaBoostClassifier

from sklearn.metrics import mean_squared_error
```

# Trees and Ensemble Methods
## Live Demos
## Decision Trees

```python
iris_data, iris_labels = load_iris().data, load_iris().target
```

```python
iris_data_train, iris_data_test, iris_labels_train, iris_labels_test = train_test_split(
    iris_data, iris_labels, train_size = 0.8, stratify = iris_labels)
```

```python
iris_data_train.shape, iris_labels_train.shape
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(iris_data_train, iris_labels_train)
```

```python
tree.score(iris_data_train, iris_labels_train)
```

```python
tree.score(iris_data_test, iris_labels_test)
```

```python
tree.get_depth()
```

```python
shallow_tree = DecisionTreeClassifier(max_depth = 3)
```

```python
shallow_tree.fit(iris_data_train, iris_labels_train)
```

```python
shallow_tree.score(iris_data_train, iris_labels_train)
```

```python
shallow_tree.score(iris_data_test, iris_labels_test)
```

```python
tree.feature_importances_
```

```python
shallow_tree.feature_importances_
```

```python
plt.figure(figsize = (10, 10))
plot_tree(tree)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
plot_tree(shallow_tree)
plt.show()
```

```python
forest = RandomForestClassifier(n_estimators = 10)
```

```python
forest.fit(iris_data_train, iris_labels_train)
```

```python
forest.score(iris_data_train, iris_labels_train)
```

```python
forest.score(iris_data_test, iris_labels_test)
```

```python
forest = RandomForestClassifier(n_estimators = 4, max_depth = 2)
forest.fit(iris_data_train, iris_labels_train)
print(forest.score(iris_data_train, iris_labels_train))
print(forest.score(iris_data_test, iris_labels_test))
```

```python
forest.feature_importances_
```

```python
boston_data, boston_prices = load_boston().data, load_boston().target
```

```python
plt.hist(boston_prices, bins = "fd")
plt.show()
```

```python
boston_data_train, boston_data_test, boston_prices_train, boston_prices_test = train_test_split(
    boston_data, boston_prices, train_size = 0.75)
```

```python
plt.hist(boston_prices_train, bins = "fd")
plt.hist(boston_prices_test, bins = "fd")
plt.show()
```

## Decision Forests

```python
forest_regressor = RandomForestRegressor(n_estimators = 5, max_depth = 5)
forest_regressor.fit(boston_data_train, boston_prices_train)
```

```python
forest_regressor.score(boston_data_train, boston_prices_train) #R-squared
```

```python
forest_regressor.score(boston_data_test, boston_prices_test) #R-squared
```

```python
np.sqrt(mean_squared_error(boston_prices_train, forest_regressor.predict(boston_data_train)))
```

```python
2.615076431677983 / boston_prices_train.mean() # %относителна грешка
```

```python
np.sqrt(mean_squared_error(boston_prices_test, forest_regressor.predict(boston_data_test)))
```

```python
 3.232190923621672 / boston_prices_test.mean() # %относителна грешка
```

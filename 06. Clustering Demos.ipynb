```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_moons, make_circles

from sklearn.cluster import KMeans, AgglomerativeClustering

from scipy.cluster.hierarchy import linkage, dendrogram
```

# Clustering
## Live Demos

```python
blobs, blob_clusters = make_blobs(n_samples = 200)
```

```python
plt.scatter(blobs[:, 0], blobs[:, 1], c = blob_clusters)
plt.show()
```

```python
blob_clusters
```

```python
k_means = KMeans(n_clusters = 3, init = "random")
```

```python
k_means.fit(blobs)
```

```python
predicted_blob_clusters = k_means.predict(blobs)
```

```python
plt.scatter(blobs[:, 0], blobs[:, 1], c = predicted_blob_clusters)
plt.show()
```

```python
blobs, blob_clusters = make_blobs(n_samples = 200, centers = [(0, 0.5), (0.5, 0), (0, 0)])
```

```python
k_means = KMeans(n_clusters = 3, init = "random")
k_means.fit(blobs)
predicted_blob_clusters = k_means.predict(blobs)
```

```python
plt.scatter(blobs[:, 0], blobs[:, 1], c = predicted_blob_clusters)
plt.show()
```

```python
blobs, blob_clusters = make_blobs(n_samples = 200, cluster_std = 2.5)
plt.scatter(blobs[:, 0], blobs[:, 1], c = blob_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters = 3, init = "random")
k_means.fit(blobs)
predicted_blob_clusters = k_means.predict(blobs)
plt.scatter(blobs[:, 0], blobs[:, 1], c = predicted_blob_clusters)
plt.show()
```

```python
blobs, blob_clusters = make_blobs(n_samples = 200, cluster_std = [(0.5, 5)] * 3)

plt.gca().set_aspect("equal")
plt.scatter(blobs[:, 0], blobs[:, 1], c = blob_clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters = 3)
k_means.fit(blobs)
predicted_blob_clusters = k_means.predict(blobs)

plt.gca().set_aspect("equal")
plt.scatter(blobs[:, 0], blobs[:, 1], c = predicted_blob_clusters)
plt.show()
```

```python
moons, moon_clusters = make_moons(n_samples = 200, noise = 0.07)
```

```python
plt.scatter(moons[:, 0], moons[:, 1], c = moon_clusters)
plt.show()
```

```python
k_means_moons = KMeans(n_clusters = 2)
k_means_moons.fit(moons)
predicted_moon_clusters = k_means_moons.predict(moons)

plt.scatter(moons[:, 0], moons[:, 1], c = predicted_moon_clusters)
plt.show()
```

```python
k_means_moons.cluster_centers_
```

```python
aggl_clustering = AgglomerativeClustering(n_clusters = 3)
```

```python
blobs, blob_clusters = make_blobs(n_samples = 200)
plt.gca().set_aspect("equal")
plt.scatter(blobs[:, 0], blobs[:, 1], c = blob_clusters)
plt.show()
```

```python
aggl_clustering.fit(blobs)
```

```python
plt.gca().set_aspect("equal")
plt.scatter(blobs[:, 0], blobs[:, 1], c = aggl_clustering.labels_)
plt.show()
```

```python
plt.figure(figsize = (20, 15))

row_clusters = linkage(blobs, method = "complete", metric = "euclidean")
dendrogram(row_clusters, labels = blob_clusters)
plt.show()
```

```python
moons, moon_labels = make_moons(noise = 0.1)

aggl_clustering = AgglomerativeClustering(n_clusters = 2, linkage = "single")
aggl_clustering.fit(moons)

plt.scatter(moons[:, 0], moons[:, 1], c = aggl_clustering.labels_)
plt.show()
```
